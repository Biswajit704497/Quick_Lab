```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
```

```python
# x = df[['Age','Sex','Blood Pressure High','Blood Pressure Low','Cholesterol','ArithmeticError','Heart Rate', 'Smoking', 'Alcohol Consumption','Exercise Hours Per Week', 'Diet','Diabetes','Family History','Previous Heart Problems','Medication Use','Stress Level']]
x = df[['Age','Sex','Blood Pressure High','Blood Pressure Low','Cholesterol','Heart Rate', 'Smoking', 'Alcohol Consumption','Exercise Hours Per Week', 'Diet','Diabetes','Family History','Previous Heart Problems','Medication Use','Stress Level']]
y = df.iloc[:,-1]
print(x.columns)
```

```python
scale = StandardScaler()
x_scale = scale.fit_transform(x) 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scale, y, test_size=0.3, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.95)
x_pca = pca.fit_transform(x)
```

```python
x_train_pca, x_test_pca, y_train, y_test = train_test_split(x_pca, y, test_size=0.3, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_pca = RandomForestClassifier()
model_pca.fit(x_train_pca, y_train)
model_pca.score(x_test_pca, y_test)
```
