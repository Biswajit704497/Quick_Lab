### Method 1 : load ecg signal


```python
def load_ecg(path_list, path):
    signals = []
    for file in path_list:
        signal, meta = wfdb.rdsamp(path+file)
        signals.append(signal)
    signals = np.array(signals)

    return signals
```

```python
file_list = clean_path_list 
chunk_size = 1000
save_folder = "ecg_chunks"
os.makedirs(save_folder, exist_ok=True)
```

```python
file_list[:3]
```

```python
pc_path = "C:/Users/ytsub/Downloads/dataset/mimic-iv-ecg-diagnostic-electrocardiogram-matched-subset-1.0/mimic-iv-ecg-diagnostic-electrocardiogram-matched-subset-1.0/"

for i in range(0, len(file_list), chunk_size):
    
    chunk_files = file_list[i:i+chunk_size]
    data = []

    for file in chunk_files:
        signal = wfdb.rdsamp(pc_path+file)[0]
        data.append(signal)

    X_chunk = np.array(data)

    # Save chunk
    np.save(f"{save_folder}/chunk_{i}.npy", X_chunk)
    
    print(f"Saved chunk {i}")

    # FREE MEMORY (VERY IMPORTANT)
    del data
    del X_chunk
    gc.collect()
```

```python
keywords = [
    "MI",
    "INFARCT",
    "ST ELEVATION",
    "ISCHEMIA"
]

def get_label(text):
    text = text.upper()
    for kw in keywords:
        if kw in text:
            return 1   # Heart attack / abnormal
    return 0           # Normal

df['label'] = df['full_report'].apply(get_label)

y = df['label'].values
```
