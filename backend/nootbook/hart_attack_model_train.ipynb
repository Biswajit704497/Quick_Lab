```python
df =df.drop('Patient ID', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
catagory_column = ['Sex','Blood Pressure','Diet','Country','Continent','Hemisphere']
for col in catagory_column:
    df[col] = le.fit_transform(df[col])
```

```python
df.head(3)
```

```python
x = df.drop("Heart Attack Risk", axis=1)
y = df['Heart Attack Risk']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.30,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

#Decision Tree Regresion 
tree_Regression=DecisionTreeClassifier()
tree_Regression.fit(X_train,y_train)
tree_Regression.score(X_test,y_test)
```
